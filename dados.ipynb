# Importar dados e pacotes

```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df.info()
```

```python
df['sold_date'] = pd.to_datetime(df['sold_date'])
```

```python
df['street'] = df['street'].fillna('-')
df['city'] = df['city'].fillna('-')
df['zip_code'] = df['zip_code'].fillna(df['zip_code'].median())
df['bed'] = df['bed'].fillna(df['bed'].median())
df['house_size'] = df['house_size'].fillna(df['house_size'].median())
df['bath'] = df['bath'].fillna(df['bath'].median())
df['acre_lot'] = df['acre_lot'].fillna(df['acre_lot'].median())
for i in ['bed','bath','acre_lot','zip_code','house_size']:
    df[f'{i}'] = df[f'{i}'].astype('float')
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.drop(columns=['full_address','street','sold_date'])
```

```python
from pycaret.regression import *
s = setup(df, target = 'price',normalize=True,remove_outliers=True)
```

```python
best = compare_models()
```
